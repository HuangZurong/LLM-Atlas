# Supervised Fine-tuning (SFT) Workflow

This notebook serves as a consolidated guide for the training pipeline, from data preparation to evaluation.

## 1. Data Preparation
Refer to the Data Strategy section for cleaning and mixture. Use `datasets` library for efficient loading.

## 2. Model Initialization
Loading weights, expanding vocabulary, and applying quantization (BitsAndBytes).

## 3. Training Setup
Setting up the `Trainer` class, learning rate schedules (Cosine), and logging (WandB/TensorBoard).

### Common SFT Training Schemes:
Depending on your resource constraints and model requirements, choose from these schemes:
1. **Scheme A**: Base Model + Domain SFT.
2. **Scheme B**: Base Model + Domain Continue Pre-train + Domain SFT.
3. **Scheme C**: Base Model + Domain Continue Pre-train + General SFT + Domain SFT.
4. **Scheme D**: Chat Model + Domain SFT.
5. **Scheme E**: Chat Model + Domain Continue Pre-train + Domain SFT.

## 4. Evaluation & Validation
Monitoring training loss vs. validation loss. Running benchmarks like C-Eval or HumanEval.
